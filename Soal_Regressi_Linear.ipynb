```python
import pandas as pd #UNTUK MANIPULASI DATA

path = '/content/drive/MyDrive/CSV/indonesia - Sheet1 (1).csv' # PATH ATAU PEMANGGILAN

df = pd.read_csv(path)#DF : DATAFRAME

df.describe()
```

```python
df
```

```python
import matplotlib.pyplot as plt #VISUALISASI DATA
```

```python
df.plot(kind='scatter', x='Life expectancy', y='percentage expenditure') #UNTUK MEMANGGI VARIABEL

plt.title('indeks kesejahteraan warga')
plt.xlabel('Life expectancy')
plt.ylabel('percentage expenditure')
plt.xlim(66, 82)
plt.ylim(0, 16)
plt.grid(True)
plt.show()
```

```python
df.plot(kind='scatter', x='Population', y='GDP') #UNTUK MEMANGGI VARIABEL

plt.title('indeks kesejahteraan warga')
plt.xlabel('Population')
plt.ylabel('GDP')
plt.xlim(0, 150)
plt.ylim(0, 16)
plt.grid(True)
plt.show()
```
